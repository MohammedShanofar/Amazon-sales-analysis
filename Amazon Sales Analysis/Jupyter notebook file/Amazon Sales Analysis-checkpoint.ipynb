```python
#importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('amazon.csv')
df
```

```python
#view first Six rows
df.head(6)
```

```python
df.shape
```

```python
df.dtypes
```

```python
#replacing string as int in actual_price
df['actual_price'] = df['actual_price'].str.replace('₹','')
df['actual_price'] = df['actual_price'].str.replace(',','').astype('float64')
df.head(6)

```

```python
#replacing string as int discounted_price
df['discounted_price'] = df['discounted_price'].str.replace('₹','')
df['discounted_price'] = df['discounted_price'].str.replace(',','').astype('float64')
df.head(6)
```

```python
#removing % from discount_precentage
df['discount_percentage'] = df['discount_percentage'].str.replace('%','').astype('float64')
df.head(6)
```

```python
#removing , from rating_count
df['rating_count'] = df['rating_count'].str.replace(',','').astype('float64')
df.head(6)
```

```python
#find the total values in 	rating
df['rating'].value_counts()
```

```python
df.duplicated()
```

```python
#Replacing | to 3.5 from rating 
df['rating'] = df['rating'].str.replace('|','3.5').astype('float64')
df['rating'].value_counts()
```

```python
df.dtypes
```

```python
df.isnull().sum()
```

```python
cat_split=df['category'].str.split('|', expand = True)
cat_split
```

```python
cat_split = cat_split.rename(columns={0:'main category 1',1:'sub category 1',2:'sub category 2'})
cat_split
```

```python
df['main category 1'] = cat_split['main category 1']
df['sub category 1'] = cat_split['sub category 1']
df['sub category 2'] = cat_split['sub category 2']
df.drop(columns = 'category',inplace = True)
df
```

```python
df.isnull().sum()
```

```python
df.dropna(how='any')
```

```python
df['main category 1'].value_counts()
```

```python
df['main category 1'] = df['main category 1'].str.replace('&',' & ')
df['main category 1'] = df['main category 1'].str.replace('OfficeProducts','Office Products')
df['main category 1'] = df['main category 1'].str.replace('MusicalInstruments','Musical Instruments')
df['main category 1'] = df['main category 1'].str.replace('HomeImprovement','Home Improvement')
df
```

```python
df['sub category 1'].value_counts()
```

```python
df['sub category 1']=df['sub category 1'].str.replace('&',' & ')
df['sub category 1']=df['sub category 1'].str.replace(',',' , ')
df['sub category 1']=df['sub category 1'].str.replace('WearableTechnology ','Wearable Technology')
df['sub category 1']=df['sub category 1'].str.replace('NetworkingDevices','Networking Devices')
df['sub category 1']=df['sub category 1'].str.replace('OfficePaperProducts','Office Paper Products')
df['sub category 1']=df['sub category 1'].str.replace('ExternalDevices','External Devices')
df['sub category 1']=df['sub category 1'].str.replace('HomeStorage','Home Storage')
df['sub category 1']=df['sub category 1'].str.replace('HomeAudio','Home Audio')
df['sub category 1']=df['sub category 1'].str.replace('GeneralPurposeBatteries','General Purpose Batteries')
df['sub category 1']=df['sub category 1'].str.replace('BatteryChargers','Battery Chargers')
df['sub category 1']=df['sub category 1'].str.replace('CraftMaterials','Craft Materials')
df['sub category 1']=df['sub category 1'].str.replace('OfficeElectronics','Office Electronics')
df['sub category 1']=df['sub category 1'].str.replace('CarAccessories','Car Accessories')
df['sub category 1']=df['sub category 1'].str.replace('HomeMedicalSupplies','Home Medical Supplies')
df
```

```python
df['sub category 2'].value_counts()
```

```python
df['sub category 2']=df['sub category 2'].str.replace('&',' & ')
df['sub category 2']=df['sub category 2'].str.replace(',',' , ')
df['sub category 2']=df['sub category 2'].str.replace('SmallKitchenAppliances','Small Kitchen Appliances')
df['sub category 2']=df['sub category 2'].str.replace('MobileAccessories','Mobile Accessories')
df
```

```python
#Task: Identify and list the top-rated products based on user ratings.
plt.figure(figsize=(10, 6))
sns.histplot(df['discounted_price'], bins=20, kde=True, color='blue', label='Discounted Price')
sns.histplot(df['actual_price'], bins=20, kde=True, color='red', label='Actual Price')

plt.title('Distribution of Product Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

```python
# Examine the distribution of discount percentages across different products.
plt.figure(figsize=(12, 6))
sns.histplot(df['discount_percentage'], bins=20, kde=True, color='blue')

plt.title('Distribution of Discount Percentages')
plt.xlabel('Discount Percentage')
plt.ylabel('Frequency')
plt.show()
```

```python
#Explore the distribution and relevance of product image links.
df[['discounted_price', 'actual_price']] = df[['discounted_price', 'actual_price']].astype(float)

# Grouped bar chart for Discounted Price vs. Actual Price
plt.figure(figsize=(12, 6))
df[['discounted_price', 'actual_price']].mean().plot(kind='bar', color=['skyblue', 'orange'])
plt.title('Discounted Price vs. Actual Price Comparison')
plt.xlabel('Price Type')
plt.ylabel('Average Price')
plt.xticks(rotation=0)  # Keep x-axis labels horizontal
plt.legend(['Discounted Price', 'Actual Price'])
plt.show()
```

```python
# Investigate the relationship between user ratings and the discount percentage offered.
plt.figure(figsize=(10, 10))

# Group by rating and calculate the average discount percentage
average_discount_by_rating = df.groupby('rating')['discount_percentage'].mean()

# Plot the bar chart
sns.histplot(x=average_discount_by_rating.index,color='red') 
sns.histplot(y=average_discount_by_rating.values, color='green')

# Add labels and title
plt.title('Average Discount Percentage by User Rating')
plt.xlabel('Average Discount Percentage')
plt.ylabel('User Rating')

plt.show()
```
